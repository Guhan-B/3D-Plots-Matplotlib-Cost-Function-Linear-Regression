Function is f(x,y) = sin(x)cos(y)

On the domain -5 <= x <= 5 -5 <= y <= 5

```python
import numpy as np

def f(x,y):
    return np.sin(x) * np.sin(y)

x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

# Contour Plot

```python
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

plt.contour(X, Y, Z, 20, cmap = cm.RdYlBu)
plt.show()
```

# Surface Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize = (12,8))
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, Z, cmap = cm.RdYlBu)
plt.show()
```

# Visualising Cost Function of Linear Regression

```python
import numpy as np
import pandas as pd
from matplotlib import cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
theta_1 = np.linspace(-5, 5, 100)
theta_2 = np.linspace(-5000, 5000, 100)

THETA_1, THETA_2 = np.meshgrid(theta_1, theta_2)
```

```python
data = pd.read_csv("./data.csv")
x = data['x']
y = data['y']
```

```python
def cost_function(THETA_1, THETA_2, x, y, n, data_size):
    cost = np.zeros((n, n))
    
    for i in range(data_size):
        cost += (THETA_1 + x[i] * THETA_2 - y[i]) ** 2
    
    cost = cost / (2 * data_size)
    
    return cost
```

```python
Z = cost_function(THETA_1, THETA_2, x, y, 100, data.shape[0])
```

```python
plt.contour(THETA_1, THETA_2, Z, 20, cmap = cm.RdYlBu)
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
ax = fig.gca(projection = '3d')
ax.plot_surface(THETA_1, THETA_2, Z, cmap = cm.RdYlBu)
plt.show()
```
